```python
pop= stats.expon.rvs(loc=18, scale=1/20, size= 1500)
samp_1= stats.expon.rvs(loc=18, scale=1/20, size= 60) 
samp_2= stats.expon.rvs(loc=18, scale=1/20, size= 60) 
```

```python
ttest_1samp(a=samp_1,popmean=pop.mean())
```

```python
ttest_1samp(a=samp_2,popmean=pop.mean())
```

```python
CI= stats.ttest_1samp(samp_1,popmean=pop.mean()).confidence_interval(confidence_level=0.95)
CI
```

```python
t_stat, p_value= ttest_1samp(a=samp_1, popmean= pop.mean())
if p_value < 0.05:
  print("We reject the null hypothesis")
else:
  print("We fail to rejet the null hypothesis")
```

```python
t_stat, p_value= ttest_1samp(a=samp_2, popmean= pop.mean())
if p_value < 0.05:
  print("We reject the null hypothesis")
else:
  print("We fail to rejet the null hypothesis")
```

In both of the graphs you can see that the data cluster is not linear due to it starting at negative 1 and curving up. Whereas if it was linear the data cluster would be closer to the center linear line. After completing the ttest we can see that both samp_1 and samp_2's pvalue is greater than 0.05. Therefor We fail to reject the null hypothesis of samp_1 and samp_2.

We will fail to reject the given hypothesis for samp_1 and samp_2.
