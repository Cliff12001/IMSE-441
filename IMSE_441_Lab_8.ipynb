c.

```python
power_analysis = TTestIndPower()
power = power_analysis.power(effect_size = 0, alpha = 0.05, nobs1= 25, alternative = 'smaller')#nobs1= sample size
# Print results
print('The power for the hypothesis testing is', round(power, 2))
```

d.

```python
sample_size = power_analysis.solve_power(effect_size = 0.4, alpha = 0.05, power = 0.9, alternative = 'larger')
print('The sample size needed for each group is', round(sample_size))
```
