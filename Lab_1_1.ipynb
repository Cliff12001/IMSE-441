```python
import pandas as pd
Lab=pd.read_csv('/content/drive/MyDrive/Lab_1_1.csv')
Lab
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(25,15))
plt.plot(Lab['Film'],Lab['Total Gross in K'], label='Total Gross in K')
plt.plot(Lab['Film'],Lab['Number of cinemas'], color='red', label= 'Number of cinemas')
plt.title('Net Gross vs Films' )
plt.xlabel('Films')
plt.ylabel('Gross in K')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.bar(Lab['Film'],Lab['Weekend in K'], label='Weekend in K')
plt.bar(Lab['Film'],Lab['Weeks on release'], color='red',bottom=Lab['Weekend in K'], label= 'Weeks on release')
plt.title('Films Gross vs. Weeks on release' )
plt.xlabel('Film')
plt.ylabel('Weekend Gross')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([BFI,Dreamz, DG,Ahimsa,SonyPictures, Disney, Universal])
plt.title('Distributor vs Total Gross Income' )
plt.xlabel('Distributor')
plt.ylabel('Gross Income')
plt.show()
```

```python
Type=['BFI','Dreamz', 'DG','Ahimsa','SonyPictures', 'Disney', 'Universal']
Profit=([sum(BFI),sum(Dreamz), sum(DG),sum(Ahimsa),sum(SonyPictures), sum(Disney), sum(Universal)])
plt.figure(figsize=(7,7))
plt.pie(Profit,labels=Type,autopct='%0.2f%%', explode=[0,0.1,0,0,0,0,0])
plt.title('Film Distibuter vs Distributers')
plt.show()
```
